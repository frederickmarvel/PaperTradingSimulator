```python
import pandas as pd
import requests
import numpy as np
import json
import sqlite3
```

```python
conn = sqlite3.connect('BackTest.db')
c = conn.cursor()

# Create the table
c.execute('''
CREATE TABLE IF NOT EXISTS trend_data (
    timestamp INTEGER PRIMARY KEY,
    index_level REAL,
    trend_score INTEGER
)
''')
```

```python
xbx_url = "https://cdi-website-three.vercel.app/indices/api/v1/trend-indicator?symbol=xbx&history=true"
response = requests.get(xbx_url)
data = response.json()['data']
print(data)
```

```python
# Insert data
for item in data:
    c.execute('''
    INSERT INTO trend_data (timestamp, index_level, trend_score) VALUES (?, ?, ?)
    ON CONFLICT(timestamp) DO UPDATE SET
    index_level=excluded.index_level,
    trend_score=excluded.trend_score
    ''', (item['timestamp'], item['index_level'], item['trend_score']))

# Commit changes and close connection
conn.commit()
conn.close()
```
